### 1. Create our `selected_features` that is the list of the columns we are going to use as our `X` data. 

```python
# DOING THIS AGAIN JUST FOR PRACTICE 

# Grabs all of the columns 
selected_features = df.columns
print(type(selected_features))

# Convert the all columns object into a regular list
selected_features = list(selected_features)

# Sanity check that it is now just a list.
print(type(selected_features))

# Remove the label column from the list
# This happnes 'in place'
selected_features.remove('label')
```

### 2. Define our `X` and `y`

```python
# 2. Define our `X` and `y` data. 

X = df[selected_features]

y = df['label']
```

```python
print(df.shape)
df.label.value_counts()
```

### 3. Train-test-split our `X` and `y` data

```python
# 3. Train-test-split our `X` and `y` data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25)
```

# 3. Build and train our model
1. Initalize an empty Support Vector Classifier model.

```python
# 1. Initalize an empty Support Vector Classifier model.
from sklearn import svm

# Initalize our Support Vector Classifier
model = svm.SVC()
```

```python
# 2. Fit that model with our training data. 
model.fit(X_train, y_train)
```

### Build a function thats input is an unfitted model, X, and y data, and runs the whole pipeline and prints a classification report and confusion matrix. 

```python
### Build a function thats input is an empty model, X, and y data, and runs the whole pipeline and prints a classification report and confusion matrix. 
def build_and_eval_model(model, X, y, random_state=23):
    print("Running report for model", model)

    X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=random_state)
    
    model.fit(X_train, y_train)

    accuracy = model.score(X_test, y_test)
    print("\n\nAccuracy: %f" %accuracy)

    y_pred = model.predict(X_test)

    print(metrics.classification_report(y_test, y_pred))

    fig, ax = plt.subplots(figsize=(16,16))
    
#     I had a bit of fun making them look different from each other
    if(str(model) == str(LogisticRegression())):
        metrics.plot_confusion_matrix(estimator=model,
                             X=X_test,
                             y_true=y_test,
                             ax=ax,
                            cmap = "cividis")
        
    if(str(model) == str(RandomForestClassifier())):
        metrics.plot_confusion_matrix(estimator=model,
                             X=X_test,
                             y_true=y_test,
                             ax=ax,
                            cmap = "magma")
        
    if(str(model) == str(svm.LinearSVC())):
        metrics.plot_confusion_matrix(estimator=model,
                             X=X_test,
                             y_true=y_test,
                             ax=ax,
                            cmap = "plasma")
        
    if(str(model) == str(MultinomialNB())):
        metrics.plot_confusion_matrix(estimator=model,
                             X=X_test,
                             y_true=y_test,
                             ax=ax,
                            cmap = "inferno")
        
    if(str(model) == str(svm.SVC())):
        metrics.plot_confusion_matrix(estimator=model,
                             X=X_test,
                             y_true=y_test,
                             ax=ax,
                            cmap = "viridis")
        
    return model

# lr = LogisticRegression()
# build_and_eval_model(lr, X, y)
```

## Run LogisticRegression, RandomForest, and Multinomial Naive Bayes through the function you just built and compare the results. 
1. Which classifier did the best, and which classifier did the worst. 

```python
# LogisticRegression
lr = LogisticRegression()
build_and_eval_model(lr, X, y)
```

```python
# RandomForest
rf = RandomForestClassifier()
build_and_eval_model(rf,X,y)
```

```python
# MultinomialNB
mnb = MultinomialNB()
build_and_eval_model(mnb, X, y)
```

```python
lsvc = svm.LinearSVC()
build_and_eval_model(lsvc, X, y)
```

```python
model = svm.SVC()
build_and_eval_model(model, X, y)
```
